# P-Value Analysis by year and class level

```python
# function for Fisher's X² test
def test(arr, y):
  noZero = arr[arr>0]
  eps = noZero.min() / 10
  arr_safe = np.where(arr==0, eps, arr)
  X = -2 * np.sum(np.log(arr_safe))
  df_arr = 2 * len(arr_safe)
  p_arr = chi2.sf(X, df_arr)

  print(f"{y} Fisher’s X² = {X:.3f}, df = {df_arr}, combined p ≈ {p_arr:.3e} \n")
  return

test(pAllYear, "All")
test(p_2020, 2020)
test(p_2021, 2021)
test(p_2022, 2022)
test(p_2023, 2023)
test(p_2024, 2024)

test(honorsClassesP, "Honors")
test(standdardClassesP, "Standard")
test(apClassesP, "AP")
```

# Cohen's w calculation by year and by class level

```python
def averageW(CohensWs, N, StartText):
  weighted_w = (N * CohensWs).sum() / N.sum()
  print(f"{StartText} Sample-size–weighted average Cohen's w: {weighted_w:.4f} \n")
  return
print("Cohen's w by year:")
averageW(ws, Ns, "All")
averageW(w_2020, Ns_2020, 2020)
averageW(w_2021, Ns_2021, 2021)
averageW(w_2022, Ns_2022, 2022)
averageW(w_2023, Ns_2023, 2023)
averageW(w_2024, Ns_2024, 2024)
print("Cohen's w by class level:")

averageW(standardClassesW, standardClassesN, "Standard")
averageW(honorsClassesW, honorsClassesN, "Honors")
averageW(apClassesW, apClassesN, "AP")


```
